```python
# Import Packages

import pandas as pd
import numpy as np
import altair as alt
```

```python
df = pd.read_csv('./kaggle_survey_2021_responses.csv')
df.head()
```

```python
# Ages
def ordered_bar(data, question, size=[500,300]):
    title = data[question][0]
    
    q_data = df.iloc[1:].groupby(question).count()['Time from Start to Finish (seconds)'].reset_index()
    
    base = alt.Chart(q_data)
    hist = base.mark_bar(color='#CB6CE6').encode(
        x=alt.X('{}:N'.format(question), title=None),
        y=alt.Y('Time from Start to Finish (seconds):Q', title = 'Count of Participants')
    ).properties(
        title=title,
        width=size[0],
        height=size[1]
    )
    
    return hist
```

```python
ordered_bar(df, 'Q1')
```

```python
ordered_bar(df, 'Q2')
```

```python
ordered_bar(df, 'Q3', size=[800,400])
```

```python
ordered_bar(df, 'Q4')
```

```python
ordered_bar(df, 'Q5')
```

```python
ordered_bar(df, 'Q6')
```
